```python
import matplotlib.pyplot as plt
import numpy as np

n = 12
X = np.arange(n)
X
```

```python
Y1 = (1-X/float(n)) * np.random.uniform(0.5, 1.0 , n)
Y1
```

函数原型：  numpy.random.uniform(low,high,size)

功能：从一个均匀分布[low,high)中随机采样，注意定义域是左闭右开，即包含low，不包含high.

```python
Y2 = (1-X/float(n)) * np.random.uniform(0.5, 1.0 , n)
Y2
```

```python
plt.bar(X, +Y1,)
plt.bar(X, -Y2,)
```

```python
plt.show()
```

```python
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor = 'white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor = 'white')
plt.show()
```

```python
plt.bar(X, +Y1, facecolor='#9999ff')
plt.bar(X, -Y2, facecolor='#ff9999')
plt.show()
```

```python
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor = 'green')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor = 'green')
plt.show()
```

```python
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor = 'green')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor = 'green')

for x,y in zip(X,Y1):
    plt.text(x, y + 0.05, '%.2f' %y, ha = 'center', va = 'bottom')  # 水平对齐， 垂直对齐

for x,y in zip(X,Y2):
    plt.text(x, -y - 0.05, '-%.2f' %y, ha = 'center', va = 'top')  # 水平对齐， 垂直对齐
    
plt.xticks(())
plt.yticks(())

plt.show()
```
