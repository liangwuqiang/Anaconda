```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x,y):
    return (1- x/2 + x**5 + y**3)* np.exp(-x**2-y**2)
```

```python
n =256
# n = 5
x = np.linspace(-3,3,n)
x
```

```python
y = np.linspace(-3,3,n)
y
```

```python
X,Y = np.meshgrid(x,y)
X
```

```python
Y
```

```python
plt.contourf(X,Y, f(X,Y),8, alpha =0.75, cmap=plt.cm.hot)
```

```python
plt.show()
```
